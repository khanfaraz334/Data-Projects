```python
import requests
#specify the url we want to scrape from
Link = "https://www.sportsadda.com/football/news/top-goalscorer-fifa-world-cup-2022-qatar-golden-boot"
#convert the web page to text
Link_text = requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
#to convert Link_text into a BeautifulSoup Object
soup = BeautifulSoup(Link_text, 'lxml')
print(soup)
```

```python
#make the indentation proper
print(soup.prettify())
```

```python
#To take a look at the title of the web page
print(soup.title)
```

```python
#Only the string not the tags
print(soup.title.string)
```

```python
#First <a></a> tag
soup.a
```

```python
soup.find_all('a')
```

```python
# Obtain information from tag <table>
table1 = soup.find('tbody')
print(table1)
```

```python
#In the table that we will fetch find the <a> </a>tags  
table1 = soup.find_all('tr')
print(table1)
```

```python
top_scorers = []
```

```python
for row in table1[1:]:  # Skip the header row
    columns = row.find_all("td")
    player_name = columns[0].text.strip()
    country = columns[1].text.strip()
    matches = int(columns[2].text.strip())
    goals = int(columns[3].text.strip())
    top_scorers.append({
        "player_name": player_name,
        "country": country,
        "matches": matches,
        "goals": goals
    })

# Print the extracted top goal scorers
for scorer in top_scorers:
    print(f"Player: {scorer['player_name']}, Country: {scorer['country']}, Matches: {scorer['matches']}, Goals: {scorer['goals']}")
```

```python
#Convert the list into a dataframe 
import pandas as pd
df = pd.DataFrame(top_scorers)
print(df)
```

```python
#To save the data into an excel file 
writer = pd.ExcelWriter('Top scorers.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='List')
writer.save()
```
